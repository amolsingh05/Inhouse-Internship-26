```python
#unpacking nested Structure
data=("Amol",(25,"Engineer"))
name, (age,profession)=data
print(f"Name:{name},Age:{age},Profession:{profession}")
```

```python
#unpacking in function Argument
def print_names(*names):
  for name in names:
    print(name)
print_names("Amol","Amrit","Rishu","Ankon","Nil")
```

```python
#combinning list with unpacking
list1=[1,2,3]
list2=[4,5,6]
combined_list=[*list1,*list2]
print(*list1)
print(f"combined List:{combined_list}\n")
```
