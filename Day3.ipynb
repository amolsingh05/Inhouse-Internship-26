```python
import numpy as np
a = np.array([1,2])
b = np.array([4,5])
c = np.matrix([[1,2],[3,4]])
d = np.matrix([[5,6],[7,8]])

print("\nC:", type(c))
print(c)
print("\nD:", type(d))
print(d)
print("\nOperation on two ndarray objects (Elementwise):")
print(a * b)
print("\nOperation on two matrix objects (Matrix multiplication):")
print(c * d)
```

```python
import numpy as np
a=np.array([[1,2],[3,4],[5,6]])
print("a=",a)
print("\n with np.transpose(a) function")
print(np.transpose(a))

print("\n with ordinary transpose")
print(a.transpose())
print("\n with a.T attribute")
print(a.T)
```

```python
#trace
a=np.array([
            [2,2,1],
            [1,3,1],
            [1,2,2]
            ])
print(a)
print("\nDiagonal elements are:")
sum=0
for i in range(len(a)):
      print(a[i][i])
      sum+=a[i][i]
print(f"\ntrace of a matrix:\n{sum}")
```

```python
import numpy as np
a=np.arange(1,10)
a.shape=(3,3)
print("a=",a)
print("Rank of Matrix is ",np.linalg.matrix_rank(a))
```

```python
import numpy as np

a = np.array([
    [2, 2, 1],
    [1, 3, 1],
    [1, 2, 2]
])

#using function
det=np.linalg.det(a)
print(det)
#inverse
inv=np.linalg.inv(a)
print("inverse: ",inv)
```

```python
import numpy as np

a = np.array([
    [2,8],[1,4]
])
inv=np.linalg.inv(a)
print("inverse: ",inv)
```

```python
#Eigen value
import numpy as np

a = np.array([
    [2, 2, 1],
    [1, 3, 1],
    [1, 2, 2]
])
w,v=np.linalg.eig(a)
print("Eigen value: ",w)
print("Eigen vector: ",v)
```
