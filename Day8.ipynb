```python
import pandas as pd

clean_data = pd.read_csv('/content/sample_data/Train.csv')  # or however you created it
```

```python
print(clean_data['Item_Fat_Content'].unique())
```

```python
print(df['Item_Fat_Content'].unique())
```

```python
#replace the repetitive values

clean_data['Item_Fat_Content'].replace({'LF':'Low Fat','reg':'Regular','low fat':'Low Fat'},inplace= True)
clean_data['Item_Fat_Content'].value_counts()
```

```python
display(clean_data.Item_Fat_Content)
```

```python
print(clean_data.columns)
```

```python
clean_data['item_category']=clean_data['Item_Identifier'].apply(lambda x: x[:2])
clean_data['item_category']=clean_data['item_category'].replace({'FD':'Food', 'NC':'Non-Consumable', 'DR':'Drinks'})
clean_data['item_category'].value_counts()
```

```python
##adding a new item
clean_data.loc[clean_data[ 'Item_Type'] == 'Non-Consumable' , 'item_fat_content'] = 'No Edible'
clean_data['Item_Fat_Content'].value_counts()
```

```python
clean_data.head()
```

```python
clean_data['Outlet_Year']=2013 - clean_data['Outlet_Establishment_Year']
clean_data['Outlet_Year']
```

```python
import matplotlib.pyplot as plt

plt.hist(clean_data['Outlet_Year'].dropna(), bins=10)
plt.xlabel('Outlet Years')
plt.ylabel('Frequency')
plt.title('Histogram of Outlet Years')
plt.show()
```

```python
# Visualize the distribution of item_fat_content
fat_content_counts = clean_data['item_fat_content'].value_counts()
plt.figure(figsize=(5, 5))
plt.bar(fat_content_counts.index, fat_content_counts.values, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel('Item Fat Content')
plt.ylabel('Count')
plt.title('Distribution of Item Fat Content')
plt.show()
```
